# Import of librairies and dataset

```
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
import re
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
! pip install pyLDAvis
import pyLDAvis
import pyLDAvis.sklearn
import matplotlib.pyplot as plt
%matplotlib inline
```

The fetch_20newsgroups dataset from sklearn will be use for this project.

```
dataset = fetch_20newsgroups(shuffle=True, random_state=1, remove=('headers', 'footers', 'quotes')) # These removing permit to keep only the texts
```

```
texts = dataset.data
```

Function which preprocess a text by removing useless information.

```
def text_preprocessing (text, min_word_length) :
  """

  text : must be a string
  min_word_length : must be a integer 

  function return a string

  """
  # put in lowercase all words
  text = text.lower()
  # remove html tags
  text = text.replace('{html}',"")
  # remove URL
  text = re.sub(r'http\S+', '', text)
  # remove numbers
  text = re.sub('[0-9]+', '', text)
  # tokenize the text and convert string to list
  tokenizer = RegexpTokenizer(r'\w+')
  text = tokenizer.tokenize(text)
  # removing of stopwords
  text = [word for word in text if len(word) > min_word_length if word not in stopwords.words('english')]
  # Lemmatization
  lemmatizer = WordNetLemmatizer()
  text=[lemmatizer.lemmatize(word) for word in text]
  # convert list to string
  text = " ".join(text)
  return text

```

Applying the function for each text.

```
texts = [text_preprocessing (text, 3) for text in texts]
```

# Vectorizer

```
vectorizer = CountVectorizer(max_features=None, stop_words='english')
data_vect = vectorizer.fit_transform(texts)
```

# Latent Dirichlet Allocation 

```
model = LatentDirichletAllocation(n_components=15, random_state=0, n_jobs=-1)

model = model.fit(data_vect)
```

```
pyLDAvis.enable_notebook()
panel = pyLDAvis.sklearn.prepare(model, data_vect, vectorizer, mds='tsne', n_jobs =-1)
panel
```
